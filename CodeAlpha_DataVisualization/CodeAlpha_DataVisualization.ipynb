```python
# ==========================================
# SUPPERSTORE DATA VISUALIZATION PROJECT
# CODEALPHA INTERNSHIP
# ==========================================

# STEP 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# To display charts properly
%matplotlib inline


# ==========================================
# STEP 2: Load Dataset
# ==========================================
df = pd.read_csv(
    r"C:\Users\shaik\Sample - Superstore.csv",
    encoding='latin1'
)

print(df.head())
# Display first 5 rows
print("First 5 Rows:")
print(df.head())


# ==========================================
# STEP 3: Dataset Information
# ==========================================

print("\nDataset Information:")
print(df.info())


# ==========================================
# STEP 4: Check Missing Values
# ==========================================

print("\nMissing Values:")
print(df.isnull().sum())


# ==========================================
# STEP 5: Remove Missing Values
# ==========================================

df.dropna(inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())


# ==========================================
# STEP 6: Check Duplicate Values
# ==========================================

print("\nDuplicate Values:")
print(df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)


# ==========================================
# STEP 7: Statistical Summary
# ==========================================

print("\nStatistical Summary:")
print(df.describe())


# ==========================================
# STEP 8: Sales by Region
# ==========================================

region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,5))

region_sales.plot(kind='bar')

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()


# ==========================================
# STEP 9: Profit by Category
# ==========================================

category_profit = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))

category_profit.plot(kind='bar')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.show()


# ==========================================
# STEP 10: Monthly Sales Trend
# ==========================================

# Convert Order Date column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group monthly sales
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

plt.figure(figsize=(10,5))

monthly_sales.plot(kind='line', marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()


# ==========================================
# STEP 11: Top 10 Products by Sales
# ==========================================

top_products = df.groupby('Sub-Category')['Sales'] \
                 .sum() \
                 .sort_values(ascending=False) \
                 .head(10)

plt.figure(figsize=(10,6))

top_products.plot(kind='bar')

plt.title('Top 10 Products by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')

plt.show()


# ==========================================
# STEP 12: Sales Distribution
# ==========================================

plt.figure(figsize=(8,5))

sns.histplot(df['Sales'], bins=30)

plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')

plt.show()


# ==========================================
# STEP 13: Correlation Heatmap
# ==========================================

plt.figure(figsize=(8,5))

sns.heatmap(
    df[['Sales', 'Profit', 'Quantity', 'Discount']].corr(),
    annot=True,
    cmap='Blues'
)

plt.title('Correlation Heatmap')

plt.show()


# ==========================================
# STEP 14: Top 10 States by Sales
# ==========================================

top_states = df.groupby('State')['Sales'] \
               .sum() \
               .sort_values(ascending=False) \
               .head(10)

plt.figure(figsize=(12,6))

top_states.plot(kind='bar')

plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Sales')

plt.show()


# ==========================================
# STEP 15: Segment-wise Sales
# ==========================================

segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(7,7))

segment_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Sales by Customer Segment')
plt.ylabel('')

plt.show()


# ==========================================
# STEP 16: Final Insights
# ==========================================

print("\n========== PROJECT INSIGHTS ==========")

print("""
1. West region generated the highest sales.

2. Technology category produced the highest profit.

3. Monthly sales show fluctuations throughout the year.

4. Certain sub-categories contribute significantly to total sales.

5. Discounts affect profit margins.

6. Consumer segment contributes major sales.

7. Data visualization helps businesses make better decisions.
""")


# ==========================================
# PROJECT COMPLETED
# ==========================================

print("\nData Visualization Project Completed Successfully!")
```
